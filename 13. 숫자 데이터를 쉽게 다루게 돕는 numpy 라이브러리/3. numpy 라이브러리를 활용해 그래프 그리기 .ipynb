numpy 라이브러리를 활용해 그래프 그리기

numpy를 사용한 모드 VS 파이썬 코드 

numpy를 사용한 코드가 더 간결하다.

반복 회수를 늘려 더 많은 개수와 숫자를 랜덤으로 추출 

numpy를 사용한 버블 차트 그리기 코드 VS 버블 차트 그리기 코드 

numpy 라이브러리르 사용하여 코드를 작성하면 for 반복문을 사용하지 않고도 많은 숫자 데이터를 생성 

np.random.rand(n) : 0~1 사이에 있는 n개의 실수(float)를 만드는 함수.

np.random.randint(a,b,n) : a이상 b 미만인 정수 n개를 만드는 함수. 

```python
# numpy를 사용한 모드 
import numpy as np
import matplotlib.pyplot as plt
dice=np.random.choice(6,10)

plt.hist(dice,bins=6) # 0,1,2,3,4,5,중 랜덤으로 추출한 숫자를 히스토그램 표현 
plt.show()
```

```python
# random라이브버리를 이용한 파이썬 코드
import matplotlib.pyplot as plt
import random 

dice=[]
for i in range(10):
    dice.append(random.randint(1,6))
    
plt.hist(dice,bins=6)
plt.show()
```

```python
import matplotlib.pyplot as ply
import numpy as np 
dice=np.random.choice(6,1000000,p=[0.1,0.2,0.3,0.2,0.1,0.1])
plt.hist(dice,bins=6)  # 0,2,3,4,5 중 랜덤으로 추출한 숫자를 히스토그램 표현 
plt.show()
```

```python
# numpy를 사용한 버블 차트 그리기 코드 
import matplotlib.pyplot as ply
import numpy as np

# 총 200번의 데이터를 추출 
x=np.random.randint(10,100,200) # 10부터 100까지의 범위 안에서 무작위로 숫자를 추출
y=np.random.randint(10,100,200) # 10부터 100까지의 범위 안에서 무작위로 숫자를 추출 
size=np.random.randint(100)*100

plt.scatter(x,y,s=size,c=x,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```

```python
# 버블 차트 그리기 코드
import matplotlib.pyplot as plt
import random

x=[]
y=[]
size=[]

for i in range(200): # 총 200개의 데이터를 추출 
    x.append(random.randint(10,100)) # 10부터 100까지의 범위 안에서 무작위로 숫자를 추출 
    y.append(random.randint(10,100)) # 10부터 100까지의 범위 안에서 무작위로 숫자를 추출 
    size=np.random.randint(100)*100
# 10부터 100 사이에 있는 정수 200개가 랜덤하게 생성 
plt.scatter(x,y,s=size,c=x,cmap='jet',alpha=0.7)
plt.colorbar()
plt.show()
```
