rate=유임승차인원/무임승차인원

(이 외에도 rate 값을 계산하는 방법은 다양할 수 있다.)

알고리즘

1) 데이터를 읽어온다.

2) 모든 역의 데이터를 바탕으로 각 역의 비율(rate)을 계산한다.

3) 비율이 가장 높은 역을 찾는다.

4) 비율이 가장 높은 역이 어디인지, 그 비율이 얼마인지 출력한다. 

```python
# 1) 데이터를 읽어온다.
import csv
f=open('subwayfee.csv',encoding='UTF8')
# 다른 운영체제에서 작성된 CSV 파일을 윈도 운영체제에서 다를 때 작성 
# UTF8은 어느 국가의 문자라도 표현할 수 있도록 만든 유니코드를 표시하는 문자 인코딩 방식 중 하나
data=csv.reader(f)
next(data)

# 2) 모든 역의 데이터를 바탕으로 각 역의 비율(rate)을 계산한다.
mx=0
rate=0
for row in data:
    for i in range(4,8):
        row[i]=int(row[i])
        rate=row[4]/row[6]
# 3) 비율이 가장 높은 역을 찾는다.        
        rate>mx:
            mx=rate
# 4) 비율이 가장 높은 역이 어디인지, 그 비율이 얼마인지 출력한다. 
    print(mx)
    
#row[6]==0인 경우가 있다 ZeroDivisionError:division by zero
```

```python
# 1) 데이터를 읽어온다.
import csv
f=open('subwayfee.csv',encoding='UTF8')
# 다른 운영체제에서 작성된 CSV 파일을 윈도 운영체제에서 다를 때 작성 
# UTF8은 어느 국가의 문자라도 표현할 수 있도록 만든 유니코드를 표시하는 문자 인코딩 방식 중 하나
data=csv.reader(f)
next(data)

# 2) 모든 역의 데이터를 바탕으로 각 역의 비율(rate)을 계산한다.
mx=0
rate=0
for row in data:
    for i in range(4,8):
        row[i]=int(row[i])
# 3) 비율이 가장 높은 역을 찾는다.    
    if(row[6]!=0): 
        rate=row[4]/row[6]
        if(rate>mx):
            mx=rate
# 4) 비율이 가장 높은 역이 어디인지, 그 비율이 얼마인지 출력한다. 
            print(row,round(mx,2))

# max값이 업데이트될 때마다 row값과 rate값을 출력 따라서, 마지막 값이 우리가 찾는 데이터 
# 문제점? 충무로역은 환슨이 가능이 가능한 역이기 떄문에 유임하차, 무임하차 값이 모두 0
```
