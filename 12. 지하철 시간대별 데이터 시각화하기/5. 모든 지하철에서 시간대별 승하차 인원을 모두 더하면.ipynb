모든 지하철역에서 시간대별 승하차 인원을 모두 더하면 

1. 데이터를 읽어온다.

2. 모든 역에 대해 시간대별 승차 인원과 하차 인원을 누적해서 더한다.

3. 시간대별 승차 인원과 하차 인원을 꺾은선 그래프로 표현한다. 

```python
# 1. 데이터를 읽어온다.
import csv
import matplotlib. pyplot as plt
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

#2. 모든 역에 대해 시간대별 승차 인원과 하차 인원을 누적해서 더한다. 
s_in=[0]*24 #승차 인원을 저장할 리스트 초기화
s_out=[0]*24 #하차 인원을 저장할 리스트 초기화

for row in data:
    row[4:]=map(int,row[4:])
    for j in range(24):
        s_in[j]+=row[2*j+4]
        s_out[j]+=row[2*j+5]
        
# 3. 시간대별 승차 인원과 하차 인원을 꺾은선 그래프로 나타낸다. 
plt.rc('font',family='Malgun Gothic') # 한글로 표현
plt.title('지하철 시간대별 승하차 인원 추이') # 제목 추가
plt.plot(s_in, label='승차') # 승차 인원을 꺾은선 그래프로 표현
plt.plot(s_out, label='하차') #하차 인원을 꺾은선 그래프로 표현 
plt.legend() # 범례 표시
plt.xticks(range(24),range(4,28))
plt.show()
# 출근시간대(7~9시)에 승차, 하차가 많고, 퇴근시간(18~19시)에 승차, 하차가 많다
#1e7은 1*10^7을 의미하고, y축에 표시된 단위가 천만 명이라는 것이다. 
```
