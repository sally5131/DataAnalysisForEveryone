```python
# 1. 데이터 불러오기
import csv
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

# 2. 시간대별로 가장 많이 타고 내리는 역 찾기
mx=[0]*24 # 시간대별 최대 승차 인원 저장 리스트 초기화
mx_station=['']*24 # 시간대별 최대 승차 인원 역 이름 저장 리스트 초기화
for row in data:
    row[4: ]=map(int,row[4:])
    for j in range(24):
        a=row[2*j+4] # j와 인덱스 변호 사이의 관계식 사용
        if a>mx[j]:
            mx[j]=a
            mx_station[j]=row[3]+'('+row[1]+')'
# 3. 시간대별로 가장 많이 타고 내리는 역이름과 인원 출력하기  
print(mx_station, mx)
# 신림역이 6번, 강남역이 14번 출력됨 24시간 중 14시간 동안 수도권 전체 지하철역 중 강남역에 가장 많은 사람이 승차한다.
```

```python
# 1. 데이터 불러오기
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

# 2. 시간대별로 가장 많이 타고 내리는 역 찾기
mx=[0]*24 # 시간대별 최대 승차 인원 저장 리스트 초기화
mx_station=['']*24 # 시간대별 최대 승차 인원 역 이름 저장 리스트 초기화
for row in data:
    row[4: ]=map(int,row[4:])
    for j in range(24):
        a=row[2*j+4] # j와 인덱스 변호 사이의 관계식 사용
        if a>mx[j]:
            mx[j]=a
            mx_station[j]=row[3]+'('+str(j+4)+'시)' # 역 이름 옆에 시간 표시
            if (j+4)>24:
                mx_station[j]=row[3]+'('+str(j+4-24)+'시)' # 역 이름 옆에 시간 표시
                
# 4. 시간대별로 가장 많이 타고 내리는 역이름과 인원을 막대그래프를 통해 표현하기
plt.rc('font',family='Malgun Gothic')
plt.bar(range(24),mx)
plt.xticks(range(24),mx_station,rotation=90) #x축에는 시간대별 1위를 차지한 역의 이름을 90도 회전하여 표현
plt.show()
```

```python
# 1. 데이터 불러오기
import csv
import matplotlib.pyplot as plt
f=open('subwaytime.csv')
data=csv.reader(f)
next(data)
next(data)

# 2. 시간대별로 가장 많이 타고 내리는 역 찾기
mx=[0]*24 # 시간대별 최대 하차 인원 저장 리스트 초기화
mx_station=['']*24 # 시간대별 최대 하차 인원 역 이름 저장 리스트 초기화
for row in data:
    row[4: ]=map(int,row[4:])
    for j in range(24):
        a=row[2*j+5] # j와 인덱스 변호 사이의 관계식 사용
        if a>mx[j]:
            mx[j]=a
            mx_station[j]=row[3]+'('+str(j+4)+'시)' # 역 이름 옆에 시간 표시
            if (j+4)>24:
                mx_station[j]=row[3]+'('+str(j+4-24)+'시)' # 역 이름 옆에 시간 표시
                
# 4. 시간대별로 가장 많이 타고 내리는 역이름과 인원을 막대그래프를 통해 표현하기
plt.rc('font',family='Malgun Gothic')
plt.bar(range(24),mx,color='b') # 막대 그래프 속성 변경 
plt.xticks(range(24),mx_station,rotation=90) #x축에는 시간대별 1위를 차지한 역의 이름을 90도 회전하여 표현
plt.show()
# 출근 시간에는 강남역이, 낮 시간에는 홍대입구역이, 퇴근 시간에는 신림역이 출력된 것을 확인할 수 있따. 
```
